```python
import pandas as pd 
import pandasql as ps 
```

```python
 df.drop(['PostTypeId','AcceptedAnswerId' ,'ParentId','DeletionDate','ViewCount','OwnerUserId','LastEditDate','LastActivityDate','Title','Tags','CommentCount','FavoriteCount','ClosedDate','CommunityOwnedDate','ContentLicense','OwnerDisplayName','LastEditorUserId','LastEditorDisplayName'],inplace=True, axis = 1) 
```

```python
q1 = """SELECT id, creationDate, score, body, answerCount  FROM ( SELECT * FROM df 
where  ( body like '%<code>%' and (body LIKE '%error%' or body LIKE '%warning%' or body LIKE '%issue%' or 
body LIKE '%exception%'or body LIKE '%fix%' or body LIKE '%problem%' or 
body LIKE '%wrong%' or body LIKE '%fail%') and answerCount > 0) LIMIT 400 ) """

final = ps.sqldf(q1,locals())
```
